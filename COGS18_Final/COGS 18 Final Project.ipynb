##### Project Description:

My project is a database with the purpose of storing a client's name, phone number, and email address. I was inspired to work on this project because of my parents; my parents' occupations require them to interact with clients and record information, and currently, they use an Excel spreadsheet to record and manage all interactions with their customers. Using an Excel spreadsheet may sometimes be confusing and cluttered, therefore I made it my goal to create a simple program with a GUI to store and display a company's list of clients.

My program takes information from a .csv file and organizes the contents in columns on the main display; the .csv file includes a client's name, phone number, and email address. Users are able to add more clients with the "Add" button, which when clicked opens a window that prompts the user to enter a client's information. After another window with a summary of the newly inputted information pops up, the main dipslay will update itself to show the new entry under the other entries.

"Remove" lets the user remove a previously added client from the database. The user must simply type in a client's name, phone number, OR email address. If the entry is found in the database, the program will warn the user whether he/she would like to remove the client. Pressing "Cancel" will end the function, and pressing "OK" will remove the client from the database. The main display will refresh to show that the action was successful. If the entry is not found, the program will state that no such entry exists.

"Search" is similar in design. "Search" allows users to look up any previously added clients in the database; they must simply type in a client's name, phone number, OR email address. If the entry is in the database, the program will open a window that displays the client's information. If a client is not found, the program will state that no such entry exists. 

Data entered/removed will not be erased when the program is terminated; data is stored in a file called "client_data.csv" in which client information is stored in rows. This file can be accessed independently, and new information can be added using a comma as a delimiter (however, you lose the aesthetics of a proper GUI).
